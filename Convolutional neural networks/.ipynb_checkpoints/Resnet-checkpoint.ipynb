```python
import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("rahmasleam/bird-speciees-dataset")

print("Path to dataset files:", path)
print(os.listdir(path))
```

```python
import torch
from torch import nn
import torch.nn.functional as f
from PIL import Image
from torchvision import transforms
```

```python
dataset_path = f"{path}/Bird Speciees Dataset"
print(os.listdir(dataset_path))

dataset = []

count_label = 0

for label in os.listdir(dataset_path):
  data_path = f"{dataset_path}/{label}"
  for example in os.listdir(data_path):
    img = Image.open(f"{data_path}/{example}").resize((400,400))
    transform = transforms.Compose([transforms.ToTensor()])
    x = 1.0 * transform(img)
    if (x.size()[0] != 3):
      continue
    y = torch.zeros(len(os.listdir(dataset_path)))
    y[count_label] = 1

    # print(x,y)

    dataset.append((x,y))

  count_label += 1
```

```python
import random

random.shuffle(dataset)
```

```python
max_w, max_h = 0, 0
for x, y in dataset:
  print(x.size())
  print(y.size())
  max_w = max(max_w, x.size()[1])
  max_h = max(max_h, x.size()[2])
```

```python
class AnimeClassifier(nn.Module):
  def __init__(self):
    super().__init__()
    self.conv0 = nn.Conv2d(3, 64, 7, 2, 3) # 400 -> 200
    self.pool1 = nn.MaxPool2d(3,2) # 98

    self.conv11 = nn.Conv2d(64, 64, 3, 1, 1)
    self.conv12 = nn.Conv2d(64, 64, 3, 1, 1)


    self.conv21 = nn.Conv2d(64, 128, 3, 2, 1)
    self.conv22 = nn.Conv2d(128, 128, 3, 1, 1)
    self.conv23 = nn.Conv2d(128, 128, 3, 1, 1)

    self.conv21 = nn.Conv2d(64, 128, 3, 2, 1)
    self.conv22 = nn.Conv2d(128, 128, 3, 1, 1)
    self.conv23 = nn.Conv2d(128, 128, 3, 1, 1)


    self.conv31 = nn.Conv2d(128, 256, 3, 2, 1)
    self.conv32 = nn.Conv2d(256, 256, 3, 1, 1)
    self.conv33 = nn.Conv2d(256, 256, 3, 1, 1)


    self.conv41 = nn.Conv2d(256, 512, 3, 2, 1)
    self.conv42 = nn.Conv2d(512, 512, 3, 1, 1)
    self.conv43 = nn.Conv2d(512, 512, 3, 1, 1)

    self.average_pool = nn.AvgPool2d(2)
    self.fc1 = nn.Linear(512 * 6 * 6, 512)
    self.fc2 = nn.Linear(512, count_label)

  def forward(self,x):
    x = f.relu(self.conv0(x))
    x = self.pool1(x)

    x11 = f.relu(self.conv11(x))
    x12 = f.relu(self.conv12(x11) + x)

    x21 = f.relu(self.conv21(x12))
    x22 = f.relu(self.conv22(x21))
    x23 = f.relu(self.conv23(x22) + x21)

    x31 = f.relu(self.conv31(x23))
    x32 = f.relu(self.conv32(x31))
    x33 = f.relu(self.conv33(x32) + x31)

    x41 = f.relu(self.conv41(x33))
    x42 = f.relu(self.conv42(x41))
    x43 = f.relu(self.conv43(x42) + x41)

    # print(x42.size())

    x = self.average_pool(x43)


    # print(x.size())
    x = torch.flatten(x)
    x = f.relu(self.fc1(x))
    x = self.fc2(x)
    return f.softmax(x, dim = 0)
```

```python
def get_label(y):
  res = 0

  for id in range(count_label):
    if y[res].item() < y[id].item():
      res = id



  return os.listdir(dataset_path)[res]
```

```python
model = AnimeClassifier()
print(get_label(model(x)))

import torch.optim as optim

optimizer = optim.SGD(model.parameters(), lr = 0.001, momentum = 0.9)
```

```python
loss_calculator = nn.CrossEntropyLoss()

for epoch in range(2):
  number_of_example = len(dataset)
  number_of_correct = 0
  print(f"=====epoch:{epoch}=====")
  for id in range(number_of_example):
    x, y = dataset[id]
    y_hat = model(x)

    loss = loss_calculator(y,y_hat)

    if get_label(y_hat) == get_label(y):
      number_of_correct += 1

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()
    print(f"{id}: {get_label(y)} - {get_label(y_hat)} || Loss: {loss.item()}")
```
